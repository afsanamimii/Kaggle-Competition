# Apply one hot encoding in cat_df and cat_df_test

```python
for x in cat_df.columns:
    one_hot= pd.get_dummies(cat_df[x])
    cat_df=cat_df.join(one_hot)
    cat_df.drop([x], axis=1,inplace=True)
```

```python
cat_df
```

```python
for x in cat_df_test.columns:
    one_hot= pd.get_dummies(cat_df_test[x])
    cat_df_test=cat_df_test.join(one_hot)
    cat_df_test.drop([x], axis=1,inplace=True)
```
