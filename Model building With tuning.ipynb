```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=df.drop(['Personal Thermal Assessment'],axis=1)   #Features(Removing ID as it is unique for all rows)
y=df['Personal Thermal Assessment']
```

# Turning Y to ordinal encoder

```python
y.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(y)
```

```python
le.classes_
```

```python
y=le.transform(y)
```

```python
val=pd.DataFrame(y)
```

```python
val.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import xgboost
```

```python
params={
 "learning_rate"    : [0.05,1] ,
 "max_depth"        : [ 3,10, 15],
 "min_child_weight" : [ 1, 3, 5],
 "gamma"            : [0.1, 0.2]  
}
```

```python
classifier=xgboost.XGBClassifier()
random_search=RandomizedSearchCV(classifier,param_distributions=params,n_iter=5,scoring='accuracy',n_jobs=-1,cv=3,verbose=3)
random_search.fit(X_train, y_train)
```

```python
random_search.best_params_ 
```

```python
classifier=xgboost.XGBClassifier({'min_child_weight': 5,
 'max_depth': 15,
 'learning_rate': 1,
 'gamma': 0.4})
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(classifier,X_train, y_train,cv=3)
```

```python
score.mean()
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix
y_pred = random_search.predict(X_test)

print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
from sklearn.metrics import roc_auc_score
print()
print(classification_report(y_test, y_pred))

plot_confusion_matrix(random_search, X_test, y_test)
```
